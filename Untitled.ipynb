```python
import pandas as pd
import numpy as np

import pymysql                        
from sqlalchemy import create_engine  

from getpass import getpass  
```

```python
password = getpass() # creating a path not to make tha password visible
```

```python
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'
engine = create_engine(connection_string)
```

```python
data = pd.read_sql_query('SELECT * FROM sakila.actor', engine) # checking if everything works
data.head()
```

```python
# How many distinct (different) actors' last names are there?
```

```python
data = pd.read_sql_query('SELECT count(distinct last_name) as distinct_names FROM actor', engine)
data.head()
```

```python
# Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
```

```python
query2 = '''
SELECT *,
CASE 
WHEN DAYOFWEEK(rental_date)= 1 OR DAYOFWEEK (rental_date) = 7 THEN "Weekend"  
ELSE "Weekday" 
END AS day_type 
FROM sakila.rental
'''

data2 = pd.read_sql_query(query2, engine)

data2
                        
```

```python
# Get all films with ARMAGEDDON in the title.
data1 = pd.read_sql_query('SELECT * FROM film WHERE title LIKE "%%ARMAGEDDON%%"', engine)
display(data1)
```

```python
# Get 10 the longest films.
```

```python
data3 = pd.read_sql_query('''SELECT title, length
FROM film
ORDER by length DESC
LIMIT 10''', engine) 
data3
```

```python
#How many films include Behind the Scenes content?
```

```python
data4 = pd.read_sql_query('''SELECT count(*) as number_of_movies
FROM film
WHERE special_features LIKE ("%%Behind the Scenes%%")''', engine)
data4
```

```python
# Which kind of movies (rating) have a mean duration of more than two hours?
```

```python
data5 = pd.read_sql_query("SELECT rating, round(AVG(length),2) as mean_duration FROM film GROUP BY rating HAVING AVG(length) > 120",engine)
data5
```
